Subset dan Encoding RIngan

```python
focus_cols = ["G3", "Dalc", "Walc", "Medu", "Fedu", "Mjob", "Fjob", "famsup", "Pstatus"]
d = df[focus_cols].copy()

# normalisasi isi string agar konsisten
for c in ["Mjob", "Fjob", "famsup", "Pstatus"]:
  d[c] = d[c].astype(str).str.strip().str.lower()

# memastikan ordinal disimpan sebagai int
for c in ["G3", "Dalc", "Walc", "Medu", "Fedu"]:
  d[c] = pd.to_numeric(d[c], errors="coerce")

d["famsup_bin"] = (d["famsup"] == "yes").astype(int)
d["pstatus_together"] = (d["Pstatus"] == "t").astype(int)

display(d.head(395))
display(d.dtypes)
```

# **Uji statistik (non-parametric & korelasi)**

Spearman correlation (ordinal)

```python
from scipy.stats import spearmanr
for col in ["Dalc","Walc","Medu","Fedu","famsup_bin"]:
    rho, p = spearmanr(d["G3"], d[col], nan_policy="omit")
    print(f"{col}: Spearman rho={rho:.3f}, p={p:.3f}")
```

Kruskal-Wallis (uji perbedaan G3 antar level Dalc/Walc)

```python
from scipy.stats import kruskal
groups = [d.loc[d["Dalc"]==lvl, "G3"].dropna() for lvl in sorted(d["Dalc"].unique())]
stat, p = kruskal(*groups)
print("Kruskal-Wallis Dalc -> stat=%.3f p=%.3f" % (stat, p))

groups_w = [d.loc[d["Walc"]==lvl, "G3"].dropna() for lvl in sorted(d["Walc"].unique())]
stat2, p2 = kruskal(*groups_w)
print("Kruskal-Wallis Walc -> stat=%.3f p=%.3f" % (stat2, p2))
```

# Regresi linear (kuantifikasi efek)

```python
import statsmodels.formula.api as smf

# Model 1: alkohol saja
m1 = smf.ols("G3 ~ Dalc + Walc", data=d).fit(cov_type="HC3")
print("Model 1: alkohol saja")
print(m1.summary().tables[1])

# Model 2: alkohol + orang tua
m2 = smf.ols("G3 ~ Dalc + Walc + Medu + Fedu + C(famsup) + C(Pstatus) + C(Mjob) + C(Fjob)", data=d).fit(cov_type="HC3")
print("\nModel 2: alkohol + orangtua")
print(m2.summary().tables[1])
```

# Statistic Report

```python
def make_summary(df_used):
    out = []
    out.append(f"Dataset: {df_used.shape[0]} rows, {df_used.shape[1]} cols")
    out.append("Mean G3: %.2f (std %.2f)" % (df_used["G3"].mean(), df_used["G3"].std()))
    out.append("Spearman correlations with G3:")
    corr = df_used[["G3","Dalc","Walc","Medu","Fedu","famsup_bin"]].corr(method="spearman")["G3"].round(3)
    out.append(str(corr))
    out.append("\nMean G3 per Dalc:\n" + str(df_used.groupby("Dalc")["G3"].mean().round(2)))
    out.append("\nMean G3 per Medu:\n" + str(df_used.groupby("Medu")["G3"].mean().round(2)))
    return "\n\n".join(out)

summary_text = make_summary(d)
print(summary_text)
```
